```python
import tensorflow as tf

BytesList = tf.train.BytesList
FloatList = tf.train.FloatList
Int64List = tf.train.Int64List
Feature = tf.train.Feature
Features = tf.train.Features
Example = tf.train.Example
```

```python
import cv2
from bs4 import BeautifulSoup
def createExample(image_file, label_file):
    img = cv2.imread(image_file)
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    image_data = tf.io.encode_jpeg(img)
#     print(image_data)
    with open(label_file, "r") as f:
        label_data = f.read()
    Bs_data = BeautifulSoup(label_data,"xml")
    bb = Bs_data.find("bndbox").get_text().split("\n")
    boundbox = [int(ele) for ele in bb if ele.strip()]
    label = Bs_data.find("name").get_text()
    return Example(
        features = Features(
            feature = {
                "image" : Feature(bytes_list = BytesList(value = [image_data.numpy()])),
                "label" : Feature(int64_list=Int64List(value=[labels.index(label.lower())])),
            }
        )
    )
```

```python
def uniqueFileName(path):
    files = os.listdir(path)
    files_set = set()
    for file in files:
        files_set.add(file[:-4])
    return list(files_set)
```

```python
def folderToFile(path, name):
    serialized_examples = []
    for file_name in uniqueFileName(path):
        print(file_name)
        example_image = createExample(os.path.join(path, file_name + ".jpg"), os.path.join(path, file_name + ".xml"))
        serialized_examples.append(example_image.SerializeToString())
    with tf.io.TFRecordWriter("{}_example.tfrecord".format(name)) as f:
        for serialized_example in serialized_examples:
            f.write(serialized_example)
            time.sleep(1)
```

```python
def to_image_dataset(filepaths, n_read_threads = 5, shuffle_buffer_size = None, 
                    n_parse_threads = 5, batch_size = 32, cache = True, repeat = False):
    dataset = tf.data.TFRecordDataset(filepaths, num_parallel_reads = n_read_threads)
    
    if cache:
        dataset = dataset.cache()
    if shuffle_buffer_size:
        dataset = dataset.shuffle(shuffle_buffer_size)
    if repeat:
        dataset = dataset.repeat()
    
    dataset = dataset.map(preprocess, num_parallel_calls=n_parse_threads)
    dataset = dataset.batch(batch_size)
    return dataset.prefetch(1)
```

```python
class ResidentialUnit(tf.keras.layers.Layer):
    def __init__(self, filters, strides = 1, activation = "relu", **kargs):
        super().__init__(**kargs)
        self.activation = tf.keras.activations.get(activation)
        self.main_layers = [
            DefaultConv2D(filters, strides = strides),
            tf.keras.layers.BatchNormalization(),
            self.activation,
            DefaultConv2D(filters),
            tf.keras.layers.BatchNormalization()
        ]
        self.skip_layers = []
        if strides > 1:
            self.skip_layers = [
                DefaultConv2D(filters, kernel_size = 1, strides = strides),
                tf.keras.layers.BatchNormalization()
            ]
    def call(self, inputs):
        Z = inputs
        for layer in self.main_layers:
            Z = layer(Z)
        skip_Z = inputs
        for layer in self.skip_layers:
            skip_Z = layer(skip_Z)
        return self.activation(Z + skip_Z)
    
    def get_config(self):
        config = super().get_config()
        return {**config, 
               "main_layer" : [keras.layers.serialize(layer) for layer in self.main_layers],
               "skip_layer" : [keras.layers.serialize(layer) for layer in self.skip_layers],
               "activation" : keras.activations.serialize(self.activation)}
```

```python
from functools import partial 
from tensorflow import keras
DefaultConv2D = partial(tf.keras.layers.Conv2D, kernel_size = 3, strides = 1, padding = "SAME", use_bias = False)

model = tf.keras.models.Sequential()
model.add(DefaultConv2D(64, kernel_size = 7, strides = 2, input_shape = [480,640,3], padding = "SAME"))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Activation("relu"))

prev_filters = 64
for filters in [64]*3:
    strides = 1 if filters == prev_filters else 2
    model.add(ResidentialUnit(filters, strides = strides))
    prev_filters = filters

model.add(tf.keras.layers.MaxPool2D(pool_size = 3, strides = 2, padding = "SAME"))
model.add(tf.keras.layers.GlobalAvgPool2D())
model.add(tf.keras.layers.Flatten())
model.add(keras.layers.Dropout(rate = 0.4))
model.add(tf.keras.layers.Dense(200, activation = "relu"))
model.add(keras.layers.Dropout(rate = 0.4))
model.add(keras.layers.Dense(4, activation = "softmax"))
```
