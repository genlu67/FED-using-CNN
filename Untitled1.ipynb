```python
def splitSetFromFile(cur_path, target_path, ratio = 0.8):
    index = np.random.permutation(len(list(pathlib.Path(cur_path).glob("*.xml"))))
    if len(index) != 0.5*len(os.listdir(cur_path)):
        print("UNMATCH ERROR in {}".format(cur_path))
    else:
        train_index = index[:int(ratio*len(index))]
        test_index = index[int(ratio*len(index)):]
        # Create Train n Test file
        TRAIN_PATH = os.path.join(target_path, "train")
        TEST_PATH = os.path.join(target_path, "test")
        if not os.path.exists(TRAIN_PATH):
            os.makedirs(TRAIN_PATH)
        if not os.path.exists(TEST_PATH):
            os.makedirs(TEST_PATH)
        
        for i in train_index:
            shutil.copy(os.path.join(cur_path,os.listdir(cur_path)[2*i]), 
                os.path.join(TRAIN_PATH,os.listdir(cur_path)[2*i]))
            shutil.copy(os.path.join(cur_path,os.listdir(cur_path)[2*i + 1]), 
                os.path.join(TRAIN_PATH,os.listdir(cur_path)[2*i + 1])) 
        for i in test_index:
            shutil.copy(os.path.join(cur_path,os.listdir(cur_path)[2*i]), 
                os.path.join(TEST_PATH,os.listdir(cur_path)[2*i]))
            shutil.copy(os.path.join(cur_path,os.listdir(cur_path)[2*i + 1]), 
                os.path.join(TEST_PATH,os.listdir(cur_path)[2*i + 1])) 
        return (TRAIN_PATH, TEST_PATH)
```

```python
import tensorflow as tf

BytesList = tf.train.BytesList
FloatList = tf.train.FloatList
Int64List = tf.train.Int64List
Feature = tf.train.Feature
Features = tf.train.Features
Example = tf.train.Example
```

```python
import cv2
def createExample(image_file, label_file):
    img = cv2.imread(image_file)
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    image_data = tf.io.encode_jpeg(img)
    print(image_data)
    with open(label_file, "r") as f:
        label_data = f.read()
    Bs_data = BeautifulSoup(label_data,"xml")
    bb = Bs_data.find("bndbox").get_text().split("\n")
    boundbox = [int(ele) for ele in bb if ele.strip()]
    label = Bs_data.find("name").get_text()
    return Example(
        features = Features(
            feature = {
                "image" : Feature(bytes_list = BytesList(value = [image_data.numpy()])),
                "label" : Feature(int64_list=Int64List(value=[labels.index(label.lower())])),
                "xmin" : Feature(int64_list=Int64List(value=[boundbox[0]])),
                "ymin" : Feature(int64_list=Int64List(value=[boundbox[1]])),
                "xmax" : Feature(int64_list=Int64List(value=[boundbox[2]])),
                "ymax" : Feature(int64_list=Int64List(value=[boundbox[3]])),
            }
        )
    )
```

```python
def preprocess(example):
    feature_description = {
#         "image" : tf.io.VarLenFeature(tf.string),
        "image": tf.io.FixedLenFeature([], tf.string, default_value=""),
        "label": tf.io.FixedLenFeature([], tf.int64, default_value = 0),
        "xmin" : tf.io.FixedLenFeature([], tf.int64, default_value = 0),
        "ymin" : tf.io.FixedLenFeature([], tf.int64, default_value = 0),
        "xmax" : tf.io.FixedLenFeature([], tf.int64, default_value = 0),
        "ymax" : tf.io.FixedLenFeature([], tf.int64, default_value = 0),    
    }
    
    example_with_image = tf.io.parse_single_example(example, feature_description)
    
    decoded_img = tf.io.decode_image(example_with_image["image"])
    print(decoded_img)
#     decoded_img = tf.reshape(decoded_img, [224, 224])
#     decoded_img = keras.applications.xception.preprocess_input(decoded_img)
    resized_image = decoded_img/255
#     final_image = keras.applications.xception.preprocess_input(resized_image)
    label = example_with_image["label"]
    boundbox = [example_with_image["xmin"]/640, 
                example_with_image["ymin"]/480,
                example_with_image["xmax"]/640,
                example_with_image["ymax"]/480]
    
     
#     label_img = example_with_image["label"].numpy()
#     decoded_img = cv2.cvtColor(decoded_img.numpy(), cv2.COLOR_BGR2RGB)
#     boundbox_img = [example_with_image[edge].numpy() for edge in ["xmin", "ymin", "xmax", "ymax"]]
#     copy_img = decoded_img.copy()
#     cv2.rectangle(copy_img, (boundbox_img[0], boundbox_img[1]), (boundbox_img[2], boundbox_img[3]), (0,0,0), 2)
#     cv2.putText(copy_img,labels[label_img],(boundbox_img[0],boundbox_img[1]), cv2.FONT_HERSHEY_COMPLEX, 1, (0,150,0), 3)
#     cv2.imshow("image_{}".format(num),
#                copy_img)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
    return resized_image, (label, boundbox)

# def preprocess_extra()
```

```python
os.listdir(THANKYOU_PATH)
def uniqueFileName(path):
    files = os.listdir(path)
    files_set = set()
    for file in files:
        files_set.add(file[:-4])
    return list(files_set)
```

```python
def folderToFile(path, name):
    serialized_examples = []
    for file_name in uniqueFileName(path):
        example_image = createExample(os.path.join(path, file_name + ".jpg"), os.path.join(path, file_name + ".xml"))
        serialized_examples.append(example_image.SerializeToString())
    with tf.io.TFRecordWriter("{}_example.tfrecord".format(name)) as f:
        for serialized_example in serialized_examples:
            f.write(serialized_example)
```

```python
def to_image_dataset(filepaths, n_read_threads = 5, shuffle_buffer_size = None, 
                    n_parse_threads = 5, batch_size = 32, cache = False, repeat = False):
    
    dataset = tf.data.TFRecordDataset(filepaths, num_parallel_reads = n_read_threads)
    
    if cache:
        dataset = dataset.cache()
    if shuffle_buffer_size:
        dataset = dataset.shuffle(shuffle_buffer_size)
    if repeat:
        dataset = dataset.repeat()
    
    dataset = dataset.map(preprocess, num_parallel_calls=n_parse_threads)
    dataset = dataset.batch(batch_size)
    return dataset.prefetch(1)
```

```python
optimizer = keras.optimizers.SGD(learning_rate=0.2, momentum=0.9, decay=0.01)

base_model = keras.applications.xception.Xception(weights = "imagenet",
                                                 include_top = False)
avg = keras.layers.GlobalAveragePooling2D()(base_model.output)
class_output = keras.layers.Dense(4, activation = "softmax")(avg)
loc_output = keras.layers.Dense(4, activation = "relu",kernel_initializer="he_normal")(avg)
model = keras.models.Model(inputs = base_model.input,
                          outputs = [class_output, loc_output])


for layer in base_model.layers:
    layer.trainable = False

model.compile(loss = ["sparse_categorical_crossentropy", "mse"],
             loss_weights = [0.8, 0.2], 
             optimizer = optimizer,
             metrics = ["accuracy"])
```
